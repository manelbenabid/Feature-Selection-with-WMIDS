```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/manelbenabid/Feature-Selection-with-WMIDS/main/datasets/Lymphoma.csv')
```

```python
df
```

```python
c = df.columns[df.eq('?').any()]
len(c)
```

```python
df = df.loc[:, ~df.isin(['?']).any()]
```

```python
df.shape
```

```python
df['class'].unique()
```

```python
y = df['class'].map({'DLBCL': 1, 'FL': 0, 'CLL':2})
```

```python
X = df.drop(columns=['class'])
X.head()
```

```python
new_columns = [i for i in range (4026)]
X.rename(columns=dict(zip(X.columns, new_columns)), inplace=True)
X.head()
```

```python
# Assuming your DataFrame is named df and has columns like "gene_1", "gene_2", ...
# Replace df with the actual name of your DataFrame

# Step 1: Compute the coefficient of variation for each column
cv_dict = {}
for column in X.columns:
    mean_value = X[column].mean()
    std_dev = X[column].std()

    # Handle division by zero when the mean is zero
    if mean_value != 0:
        coefficient_of_variation = (std_dev / mean_value) * 100
    else:
        coefficient_of_variation = 0

    cv_dict[column] = coefficient_of_variation

# Step 2: Create a dictionary and save it to a CSV file
cv_df = pd.DataFrame(list(cv_dict.items()), columns=['Gene', 'Coefficient_of_Variation'])
cv_df.to_csv('/content/drive/MyDrive/CP4/results/Lymphoma/coefficient_of_variation_results_Lymphoma.csv', index=False)

# Optional: Print the dictionary
print(cv_dict)
```

```python
from scipy.stats import pearsonr

# Assuming your dataset columns represent different genes or features
# You can calculate pairwise correlations between all features
correlation_matrix = X.corr()

correlation_matrix
```

```python
# Filter rows and columns where correlation > 0.5
filtered_corr_matrix = correlation_matrix[(correlation_matrix >= 0.8) & (correlation_matrix < 1)]

# Drop rows and columns with NaN values (correlations below 0.5)
filtered_corr_matrix = filtered_corr_matrix.dropna(axis=0, how='all')
filtered_corr_matrix = filtered_corr_matrix.dropna(axis=1, how='all')

# Print or use the filtered correlation matrix as needed
filtered_corr_matrix
```

```python
# Count correlations that are >= 0.7 and < 1
count_high_corr = ((correlation_matrix >= 0.8) & (correlation_matrix < 1)).sum().sum()

# Print the count
print("Number of correlations >= 0.5 and < 1:", count_high_corr)
```

```python
# Create a DataFrame to store the gene pairs and correlations
result_df = pd.DataFrame(columns=['first_gene', 'second_gene'])

# Iterate over the filtered correlation matrix and append data to the DataFrame
for gene1 in filtered_corr_matrix.columns:
    for gene2 in filtered_corr_matrix.index:
        correlation_value = filtered_corr_matrix.loc[gene2, gene1]
        if not np.isnan(correlation_value):
            result_df = result_df.append({'first_gene': gene1, 'second_gene': gene2}, ignore_index=True)

# Save the DataFrame to a CSV file
result_df.to_csv('/content/drive/MyDrive/CP4/results/Lymphoma/Lymphoma_gene_correlations_0.8.csv', index=False)
```
