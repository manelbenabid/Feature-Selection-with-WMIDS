## Standarizing the data

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV, train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score



# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X07, y, test_size=0.2, random_state=42, stratify=y)

# Standardize the features (important for regularization methods)
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
from sklearn.linear_model import Lasso
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.metrics import accuracy_score

# Create the Lasso model
lasso = Lasso()

# Set up a grid of hyperparameters to search
param_grid = {'alpha': [0.001, 0.01, 0.1, 1.0, 10.0]}

# Use GridSearchCV to find the best hyperparameters
grid_search = GridSearchCV(lasso, param_grid, cv=5, scoring='neg_mean_squared_error')
grid_search.fit(X_train_scaled, y_train)

# Get the best hyperparameters
best_alpha = grid_search.best_params_['alpha']

# Fit the Lasso model with the best hyperparameters on the entire training set
lasso_best = Lasso(alpha=best_alpha)
lasso_best.fit(X_train_scaled, y_train)

# Identify the selected features
selected_features_lasso_07 = X07.columns[lasso_best.coef_ != 0]

# Evaluate the model on the test set
y_pred = lasso_best.predict(X_test_scaled)
y_pred_binary = (lasso_best.predict(X_test_scaled) > 0.5).astype(int)
accuracy = accuracy_score(y_test, y_pred_binary)
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)


# Print results
print("Best alpha:", best_alpha)
print("Selected Features:", selected_features_lasso_07)
print("Test Accuracy:", accuracy)
print(f"Mean Squared Error (MSE): {mse}")
print(f"R-squared (R2): {r2}")
```

## SVM-RFE

```python
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
from sklearn.datasets import make_classification
from sklearn.model_selection import StratifiedKFold
from sklearn.pipeline import Pipeline

# Create an SVM model
svm_model = SVC(kernel='linear')

# Create a pipeline with RFE and SVM
pipeline = Pipeline([
    ('feature_selection', RFE(svm_model)),
    ('classification', svm_model)
])

# Set the range of features to select
n_features_range = list(range(20, 50))  # Adjust the range based on your dataset

# Create a parameter grid
param_grid = {
    'feature_selection__n_features_to_select': n_features_range
}

# Perform cross-validated grid search
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)  # Adjust the number of splits as needed
grid_search = GridSearchCV(pipeline, param_grid=param_grid, cv=cv, scoring='accuracy')
grid_search.fit(X_train_scaled, y_train)

# Print the results
print("Best number of features:", grid_search.best_params_['feature_selection__n_features_to_select'])
print("Best cross-validated accuracy:", grid_search.best_score_)

# Use the best number of features from the grid search
n_best_features = grid_search.best_params_['feature_selection__n_features_to_select']

# Create an SVM model with RFE using the best number of features
final_svm_model = SVC(kernel='linear')
final_rfe = RFE(final_svm_model, n_features_to_select=n_best_features)

# Fit the final model on the training set
final_rfe.fit(X_train_scaled, y_train)
# Get the selected features
selected_features_rfe_07 = X07.columns[final_rfe.support_]
# Make predictions on the test set
y_pred = final_rfe.predict(X_test_scaled)

# Compute and print the test accuracy
test_accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy:", test_accuracy)
print("Selected Features:", n_best_features)
print(f"Mean Squared Error (MSE): {mse}")
```
